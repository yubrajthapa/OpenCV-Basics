```python
# Reading video from camera
import cv2
  
  
# define a video capture object
vid = cv2.VideoCapture(2)
  
while(True):
      
    # Capture the video frame
    # by frame
    ret, frame = vid.read() # if the frame is availabe read method will return true.
    # And the frame will be saved into frame variable.
    # True or false will be stored in the ret variable
    # If frame is available ret will be true else false.
    # Display the resulting frame. 
    
    cv2.imshow('Video',frame) #Displaying the captured video.
      
    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choice
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```

```python
# Saving the captured video
import cv2

cap = cv2.VideoCapture(2)
fourcc = cv2.VideoWriter_fourcc(*'XVID') #Getting fourcc code using class called VideoWriter_fourcc
out = cv2.VideoWriter('output2.avi',fourcc,20.0,(640,480)) #20.0 is the number of frames/sec
print(cap.isOpened())                           # This is the size of frame.
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:  # if ret is true then performs following.
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # gives width of frame
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        out.write(frame)
    
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame',gray)
    
        if cv2.waitKey(1)  & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
out.release() #releasing the resources.
cv2.destroyAllWindows()
```

# Draw different Geometric shapes

```python
import numpy as np
import cv2

#img = cv2.imread('lena.jpg',1) # changing 1 to 0 makes gray colored photo.
                    #Coordinates      #Colors    #Thickness
img = np.zeros([512,512,3], np.uint8) # This gives the black image
    
img = cv2.line(img, (0,0), (255,255),  (147, 96, 44), 3)
img = cv2.arrowedLine(img, (0,255),  (255,255), (255,0,0),5)
img = cv2.rectangle(img, (384,0), (510,128),  (0,0,255),10) #When we change thickness to -1 it fills the symbols with the given color.
img = cv2.circle(img, (447,63), 63,(0,255,0),-1)

font = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
img = cv2.putText(img, 'OpenCv', (10,500), font, 4  ,(0,255,255),10,  cv2.LINE_AA)
                                #Coordinates  #fontsize          #Thickness  #linetype

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 4. Setting camera parameters in Open-CV

```python
import cv2
cap = cv2.VideoCapture(2)
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # Deafult weidth
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # default height

cap.set(3,340) #Width for frame
cap.set(4,440) #Height for the frame
#Camera will automatically take the resolution available for the default camera.

print(cap.get(3))
print(cap.get(4))


while(cap.isOpened()):
    ret,frame = cap.read()
    if ret == True:
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', gray)
        
        if cv2.waitKey(1)  &0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

# 5. Show Date and Time on Videos. 

```python
import cv2
cap = cv2.VideoCapture(2)
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # Deafult weidth
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # default height

cap.set(3,740) 
cap.set(4,740) 


print(cap.get(3))
print(cap.get(4))


while(cap.isOpened()):
    ret,frame = cap.read()
    if ret == True:
        
        
        font = cv2.FONT_HERSHEY_SIMPLEX #font for text
        text = 'width: ' +str(cap.get(3)) +' ' + 'Height: ' + str(cap.get(4))
        # text to be printed.
        
        
        frame = cv2.putText(frame, text, (10,50), font, 1,
                            (0,255,255), 2,cv2.LINE_8)
    # passing text font etc. to the frame
        cv2.imshow('Frame', frame)
        
        if cv2.waitKey(1)  &0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

# Showing date and time

```python
import cv2
import datetime
cap = cv2.VideoCapture(2)
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # Deafult weidth
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # default height



while(cap.isOpened()):
    ret,frame = cap.read()
    if ret == True:
        
        
        font = cv2.FONT_HERSHEY_SIMPLEX #font for text
        text = 'width: ' +str(cap.get(3)) +' ' + 'Height: ' + str(cap.get(4))
        # text to be printed.
        
        date = str(datetime.datetime.now())
        
        frame = cv2.putText(frame, date, (10,50), font, 1,
                            (0,255,255), 2,cv2.LINE_8)
    # passing text font etc. to the frame
        cv2.imshow('Frame', frame)
        
        if cv2.waitKey(1)  &0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
